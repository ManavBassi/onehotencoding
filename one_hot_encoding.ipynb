```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("cars.csv")
```

```python
df.head()
```

# 1. OneHotEncoding using Pandas

```python
# in this get dummies,drop_first is used for droping first dummy variable
pd.get_dummies(df,columns=['fuel','owner'],dtype='int32',drop_first=True)
```

# OneHotEncoding using sklearn

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.3,random_state=0)
x_train
```

```python
x_test
```

```python
ohe=OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

```python
x_train_new=ohe.fit_transform(x_train[['fuel','owner']])
x_train_new
```

```python
x_test_new=ohe.transform(x_test[['fuel','owner']])
x_test_new
```
